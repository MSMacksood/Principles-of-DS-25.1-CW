```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.proportion import proportions_ztest

# Load data
df = pd.read_csv('data/cw1_data.csv')

print("Data info:")
print(df.info())

results = {}

# a. Hypothesis 01: Average age of a respondent is 30years.
# One-sample t-test
t_stat_a, p_val_a = stats.ttest_1samp(df['Age'], 30)
results['a'] = {'t_stat': t_stat_a, 'p_val': p_val_a, 'mean_age': df['Age'].mean()}

# b. Hypothesis 02: There is relationship between income before the program and income after the program
# Pearson correlation
corr, p_val_b = stats.pearsonr(df['Income_Before'], df['Income_After'])
results['b'] = {'corr': corr, 'p_val': p_val_b}

# c. Hypothesis 03: More than half of the respondents completed the program
# Program_Status values: assume 1=completed
count_completed = df['Program_Status'].sum()
nobs = len(df)
stat_c, p_val_c = proportions_ztest(count=count_completed, nobs=nobs, value=0.5, alternative='larger')
results['c'] = {'count': count_completed, 'nobs': nobs, 'prop': count_completed/nobs, 'z_stat': stat_c, 'p_val': p_val_c}

# d. Hypothesis 04: Income before the program is different with respect to the level of education
# One-way ANOVA
edu_levels = [group['Income_Before'].values for name, group in df.groupby('Education')]
f_stat_d, p_val_d = stats.f_oneway(*edu_levels)
results['d'] = {'f_stat': f_stat_d, 'p_val': p_val_d}

# e. Hypothesis 05: Income before the program is different with respect to the gender
# Two-sample t-test
gender_groups = [group['Income_Before'].values for name, group in df.groupby('Gender')]
if len(gender_groups) == 2:
    t_stat_e, p_val_e = stats.ttest_ind(gender_groups[0], gender_groups[1], equal_var=False)
    results['e'] = {'t_stat': t_stat_e, 'p_val': p_val_e}

print("\nResults:")
for k, v in results.items():
    print(f"Hypothesis {k}: {v}")
```

```python
import pandas as pd
import scipy.stats as stats

# Load data
df = pd.read_csv('data/cw1_data.csv')
results = {}
```

```python
# Hypothesis 06: Income after the program is significantly greater than income before the program.
# Paired t-test (Right-tailed)
t_stat_6, p_val_6 = stats.ttest_rel(df['Income_After'], df['Income_Before'], alternative='greater')
mean_diff = (df['Income_After'] - df['Income_Before']).mean()
results['6'] = {'t_stat': t_stat_6, 'p_val': p_val_6, 'mean_diff': mean_diff}
print (results['6'])
```

```python
# Hypothesis 07: Income after the program differs between those who completed the program and those who didn't.
# Two-sample t-test
group_completed = df[df['Program_Status'] == 1]['Income_After']
group_not_completed = df[df['Program_Status'] == 0]['Income_After']
t_stat_7, p_val_7 = stats.ttest_ind(group_completed, group_not_completed, equal_var=False)
results['7'] = {'t_stat': t_stat_7, 'p_val': p_val_7, 'mean_comp': group_completed.mean(), 'mean_not_comp': group_not_completed.mean()}
print (results['7'])
```

```python
# Hypothesis 08: Program completion status is associated with marital status.
# Chi-square test of independence
contingency_table = pd.crosstab(df['Program_Status'], df['Marital_Status'])
chi2_stat, p_val_8, dof, expected = stats.chi2_contingency(contingency_table)
results['8'] = {'chi2_stat': chi2_stat, 'p_val': p_val_8}
print (results['8'])
```

```python
# Hypothesis 09: There is a relationship between Age and Income before the program.
# Pearson Correlation
corr_9, p_val_9 = stats.pearsonr(df['Age'], df['Income_Before'])
results['9'] = {'corr': corr_9, 'p_val': p_val_9}
print (results['9'])
```

```python
# Hypothesis 10: The number of household members is different depending on marital status.
# Two-sample t-test
hh_married = df[df['Marital_Status'] == 1]['Household_Members']
hh_single = df[df['Marital_Status'] == 0]['Household_Members']
t_stat_10, p_val_10 = stats.ttest_ind(hh_married, hh_single, equal_var=False)
results['10'] = {'t_stat': t_stat_10, 'p_val': p_val_10, 'mean_married': hh_married.mean(), 'mean_single': hh_single.mean()}
print(results['10'])
```

```python

print("Results for New Hypotheses:")
for k, v in results.items():
    print(f"H{k}: {v}")
```
